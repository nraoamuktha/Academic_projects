### Question 2. To read price data from the CSV file and perform data analytics

We have defined a function called **'read_price_data'** that will perfom certain tasks to analyze each dataset. The function will read the CSV file and load it to a new dataframe **'stock_hist'** and perform some statistical analysis as well as showcase the data for further analysis on our end.

The function is reusable for both datasets.

Parameters used in the function:
   - filename: The CSV file to read.
   - index_col: The column to use as the index (default is 0).
   - parse_dates: Whether to parse dates as datetime objects (default is True).
    


```python
# Define function read price data
def read_price_data(filename, index_col=0, parse_dates=True):

    # Load the csv file into a new dataframe stock_hist
    stock_hist = pd.read_csv(filename, index_col=index_col ,parse_dates=parse_dates)

    # Verify dataframe information by highlighting the DatatimeIndex
    print('Stock Information:\n')
    print(f'{stock_hist.info()}\n')

    # Check the existence of null values
    print('NULL values check:\n')
    print(f'{stock_hist.isnull().sum()}\n')

    # Showcase the summary statistics
    print('Summary statistics:\n')
    print(f'{stock_hist.describe()}\n')

    # Verify the start dates
    print('Check the first few rows:\n')
    print(f'{stock_hist.head()}\n')

    # Verify the end dates
    print('Check the last few rows:\n')
    print(f'{stock_hist.tail()}\n')

    return stock_hist
```

### Question 4. Define a function to plot and format Y-Axis

We have defined a function named **'plot_and_format_y_axis'** that plots the data from the dataframe in a line graph as well as  formats and customize the plots.

Parameters used in the function:
 - data_frame : The dataframe used to plot
 - formatter(mpl.ticker.ScalarFormatter()) : The Y axis formater which defaults to a scalar
 - aspect_ratio : Defaults the inline plots in the ratio of 12:8
 - title : Plot title
 - ylabel: Y-axis label
 - grid : To display the grid line (True)



```python
# Function to plot columns
def plot_and_format_y_axis(data_frame, formatter= mpl.ticker.ScalarFormatter() ,aspect_ratio=(12,8), title='', ylabel='', grid = True):

    pt.figure(figsize=aspect_ratio)
    for column in data_frame.columns:
        pt.plot(data_frame.index, data_frame[column], label=column)

    pt.title(title)
    pt.xlabel('Date')
    pt.ylabel(ylabel)
    pt.grid(grid)
    pt.legend()
    if formatter:
        pt.gca().yaxis.set_major_formatter(formatter)
    pt.show()
```

### Question 6. Extraction and Merging the Adj_Close columns.
We have taken the adjusted close columns from both datasets and combined them into a single dataframe using an inner join in order to compare the analysis of the Adj_close prices of the META and SPY files. Since it takes into account corporate actions like dividends and stock splits, the Adj close price is one of the most widely used metrics for stock price analysis.


```python
# Extract the Adjusted close columns of both stocks
SPY_adj = pd.DataFrame(SPY_df['Adj Close'])
META_adj = pd.DataFrame(META_df['Adj Close'])

# Rename the columns of both dataframes
SPY_adj.columns = ['SPY']
META_adj.columns = ['META']

# Merge both stocks into a single dataframe
merged_adj_close = pd.merge(SPY_adj, META_adj, how='inner', left_index=True, right_index=True)
merged_adj_close
```

```python
# Plot the adjusted close prices of securities
aspect_ratio = (12,8)
merged_adj_close.plot(figsize=aspect_ratio, title='Adjusted Close Prices', ylabel = 'Price')
```

### Question 8. Creating and Analyzing a Normalized Portfolio


```python
#To create this "normalized" portfolio is to simply divide each day's price by the initial price P0.
# This yields a portfolio with starting values of $1. You can then simply look at that portfolio, or you can multiply the
# entire portfolio by a fixed initial investment amount, say, $1,000
initial_investment = 1000
portfolioNormalized = merged_adj_close / merged_adj_close.iloc[0]
```

```python
# next, we can simply multiply this portfolio by the scalar $1,000 to get the previously constructed portfolio
portfolioAlternate = initial_investment * portfolioNormalized
```

```python
# that gives us back the previously constructed portfolio
portfolioAlternate
```

```python
#Ploting Normalized Portfolio vs Initial Investment Amount
portfolioAlternate.plot(figsize=aspect_ratio, title='Normalized Portfolio x Initial Investment Amount')
```

### Question 12 Fator to equlize risk for both securities

The standard deviations of both the securities are not the same.  
Since the standard deviation in META's security is higher, we have to multilply the lower risk security SPY with a factor to equalize risk.

To equalize the risk between these two securities, we calculate a risk equalization factor. This factor is derived by dividing the standard deviation of META's security by the standard deviation of SPY. The formula is:

```python
# Calculate the factor to equlize risk for both securities
equal_risk_factor = price_diff['META'].std()/price_diff['SPY'].std()
# Display equal risk factor
print("equal_risk_factor is:", equal_risk_factor)
```

The **equal_risk_factor** helps to quatify the ratio of risk between both the securities. It is helpful in adjustment of low risk security i.e., SPY with high risk security i.e., META.
This ratio calculation helps to ensure that the investment portfolio is in line with desired risk levels and make informed decisions about portfolio composition and risk management.

### Question 13 Create new Equal risk portfolio

To create a new risk equal investment portfolio, we have to mulitply the initial investment of SPY with the previously calculated factor.

```python
# Create a new equal risk portfolio by adjusting SPY's investment
## Clone the portfolioAlternate
equal_risk_portfolio = portfolioAlternate.copy()

## Adjust SPY's investment by multiplying by the equal risk factor
equal_risk_portfolio['SPY'] = equal_risk_portfolio['SPY'] * equal_risk_factor

#Display equal risk portfolio
equal_risk_portfolio
```

**Which security outperforms?**

Calculate the “return per unit of risk” numbers for both securities in both the original
